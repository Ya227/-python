```python
import abc
from error import InvalidEquipmentError
import logging

logger = logging.getLogger(__name__)

class EquipmentFactory:
    @staticmethod
    def create_equipment(data):
        try:
            InvalidEquipmentError.validate(data)
            equipment_type = data.pop('equipment_type')
            equipment_class = EquipmentMeta.get_equipment(equipment_type, **data)
            if not equipment_class:
                raise ValueError(f"Неизвестный тип оборудования: {equipment_type}")
            return equipment_class
        except InvalidEquipmentError as e:
            logger.error(f"Ошибка создания оборудования: {e.mes}")
            raise

class EquipmentMeta(abc.ABCMeta):
    _registry = {}
    def __new__(cls, name, bases, dct):
        new_class = super().__new__(cls, name, bases, dct)
        if name != 'Equipment' and issubclass(new_class, Equipment):
            cls._registry[name] = new_class
        return new_class

    @classmethod
    def get_equipment(cls, equipment_type, **mas):
        equipment_class = cls._registry.get(equipment_type)
        if equipment_class:
            if equipment_type == 'Camera':
                return equipment_class(
                    equipment_id=mas['equipment_id'],
                    name=mas['name'],
                    condition=mas['condition'],
                    daily_rate=mas['daily_rate'],
                    sensor_type=mas['sensor_type']
                )
            elif equipment_type == 'Lens':
                return equipment_class(
                    equipment_id=mas['equipment_id'],
                    name=mas['name'],
                    condition=mas['condition'],
                    daily_rate=mas['daily_rate'],
                    focal_length=mas['focal_length']
                )
            elif equipment_type == 'Accessory':
                return equipment_class(
                    equipment_id=mas['equipment_id'],
                    name=mas['name'],
                    condition=mas['condition'],
                    daily_rate=mas['daily_rate'],
                    compatibility=mas['compatibility']
                )
        return None

class Equipment(metaclass = EquipmentMeta):
    def __init__(self, equipment_id, name, condition, daily_rate, equipment_type, **kwargs):
        self._equipment_id = equipment_id
        self._name = name
        self._condition = condition
        self._daily_rate = daily_rate
        self._equipment_type = equipment_type
        self._is_available = True
        self.__dict__.update(kwargs)
        logger.info(f"Создано оборудовани: {self._name} (ID: {self._equipment_id})")

    @property
    def equipment_id(self):
        return self._equipment_id

    @property
    def name(self):
        return self._name

    @name.setter
    def name(self, value):
        if isinstance(value, str):
            logger.debug(f"Изменено название оборудования {self._name} на {value}")
            self._name = value
        else:
            raise ValueError("Не подходящие данные для названия")

    @property
    def condition(self):
        return self._condition

    @condition.setter
    def condition(self, value):
        if isinstance(value, str):
            logger.debug(f"Изменено состояние оборудования {self._name} на {value}")
            self._condition = value
        else:
            raise ValueError("Не подходящие данные для состояния")

    @property
    def daily_rate(self):
        return self._daily_rate

    @daily_rate.setter
    def daily_rate(self, value):
        if isinstance(value, (int, float)) and value >= 0:
            logger.debug(f"Изменена стоимость аренды за день оборудования {self._name} на {value}")
            self._daily_rate = value
        else:
            raise ValueError("Не подходящие данные для стоимости аренды за день")

    @property
    def is_available(self):
        return self._is_available

    @is_available.setter
    def is_available(self, value):
        if isinstance(value, bool):
            logger.debug(f"Изменена доступность оборудования {self._name} на {value}")
            self._is_available = value
        else:
            raise ValueError("Не подходящие данные для доступности оборудования")


    def to_dict(self):
        logger.debug(f"Преобразование оборудования {self._equipment_type} {self._equipment_id} в словарь")
        return {
            'equipment_type': self._equipment_type,
            'equipment_id': self._equipment_id,
            'name': self._name,
            'condition': self._condition,
            'daily_rate': self._daily_rate,
            'is_available': self._is_available
            **self.__dict__
        }

    @classmethod
    def from_dict(cls, data):
        logger.debug(f"Создание оборудования из словаря: {data}")
        return cls(**data)

    @abc.abstractmethod
    def calculate_rental_cost(self, days):
        pass

    def __str__(self):
        return f"Оборудование: {self._name}, Состояние: {self._condition}"

    def __lt__(self, other):
        if self._daily_rate != other._daily_rate:
            return self._daily_rate < other._daily_rate
        return self._condition < other._condition

    def __gt__(self, other):
        if self._daily_rate != other._daily_rate:
            return self._daily_rate > other._daily_rate
        return self._condition > other._condition

    def __eq__(self, other):
        return self._daily_rate == other._daily_rate and self._condition == other._condition


from goods import Camera, Lens, Accessory
```

```python
from equipment import Equipment, logger

class Camera(Equipment):
    def __init__(self, equipment_id, name, condition, daily_rate, sensor_type):
        super().__init__(equipment_id, name, condition, daily_rate, 'Camera')
        self._sensor_type = sensor_type
        logger.info(f"Создана камера: {name} (ID: {equipment_id}, Сенсор: {sensor_type})")

    @property
    def sensor_type(self):
        return self._sensor_type

    @sensor_type.setter
    def sensor_type(self, value):
        if isinstance(value, str):
            logger.debug(f"Изменен тип сенсора камеры {self._name} на {value}")
            self._sensor_type = value
        else:
            raise ValueError("Неподходящие данные для сенсора камеры")

    def calculate_rental_cost(self, days):
        if(days >= 7):
            cost = float(self._daily_rate * days) * 0.75
        else:
            cost = self._daily_rate * days
        logger.debug(f"Расчет стоимости аренды камеры {self._name}: {cost} за {days} дней")
        return cost

    def __str__(self):
        return super().__str__() + f", Тип сенсора: {self._sensor_type}"

class Lens(Equipment):
    def __init__(self, equipment_id, name, condition, daily_rate, focal_length):
        super().__init__(equipment_id, name, condition, daily_rate, 'Lens')
        self._focal_lengtf = focal_length
        logger.info(f"Создан объектив: {name} (ID: {equipment_id}), ФР: {focal_length})")

    @property
    def focal_length(self):
        return self._focal_lengtf

    @focal_length.setter
    def focal_length(self, value):
        if isinstance(value, str):
            logger.debug(f"Изменено фокусное расстояние объектива {self._name} на {value}")
            self._focal_lengtf = value
        else:
            raise ValueError("Неподходящие данные для фокусного расстояния объектива")

    def calculate_rental_cost(self, days):
        if(days >= 7):
            cost = float(self._daily_rate * days) * 0.75
        else:
            cost = self._daily_rate * days
        logger.debug(f"Расчет стоимости аренды объектива {self._name}: {cost} за {days} дней")
        return cost

    def __str__(self):
        return super().__str__() + f", Фокусное расстояние: {self._focal_lengtf}"

class Accessory(Equipment):
    def __init__(self, equipment_id, name, condition, daily_rate, compatibility):
        super().__init__(equipment_id, name, condition, daily_rate, 'Accessory')
        self._compatibility = compatibility
        logger.info(f"Создан аксессуар: {name} (ID: {equipment_id}), Совместимость: {compatibility})")


    @property
    def compatibility(self):
        return self._compatibility

    @compatibility.setter
    def compatibility(self, value):
        if isinstance(value, bool):
            logger.debug(f"Изменена совместимость аксессуара {self._name} на {value}")
            self._compatibility = value
        else:
            raise ValueError("Неподходящие данные для совместимости аксессуара")

    def calculate_rental_cost(self, days):
        if(days >= 3):
            cost = float(self._daily_rate * days) * 0.9
        else:
            cost = self._daily_rate * days
        logger.debug(f"Расчет стоимости аренды аксессуара {self._name}: {cost} за {days} дней")
        return cost

    def __str__(self):
        return super().__str__() + f", Совместимость с оборудованием: {self._compatibility}"
```

```python
from equipment import Equipment, logger
from error import PermissionDeniedError
from datetime import datetime

class Customer():
    def __init__(self, customer_id, name, email, number, role = "user"):
        self.customer_id = customer_id
        self.name = name
        self.email = email
        self.number = number
        self.role = role
        logger.info(f"Создан клиент: {self.name} (ID: {self.customer_id}, Email: {self.email})")

    def to_dict(self):
        logger.debug(f"Преобразование клиента {self.name} в словарь")
        return {
            'customer_id': self.customer_id,
            'name': self.name,
            'email': self.email,
            'number': self.number,
            'role': self.role
        }

    @classmethod
    def from_dict(cls, data):
        logger.debug(f"Создание клиента из словаря: {data}")
        return cls(**data)

    def can_rent(self):
        if self.role not in ["user", "operator", "manager", "admin"]:
            logger.warning(f"Попыткак аренды клиента без прав: {self.name}")
            raise PermissionDeniedError(f"У пользователя нет прав")
        return True

    def __str__(self):
        return f"Клиент: {self.name}, Email: {self.email}, Номер: {self.number}"

class Rental():
    def __init__(self, rental_id, customer, equipment, start_date, end_date, accessories = None):
        self.rental_id = rental_id
        self.customer_info = customer
        self.equipment = equipment
        self.start_date = start_date
        self.end_date = end_date
        self.total_cost = self.calculate_total()
        self.accessories = accessories if accessories is not None else []
        logger.info(f"Создана аренда: ID {self.rental_id}, Клиент: {self.customer_info.name}, Оборудование: {self.equipment.name}, Аксессуары: {self.accessories}")

    def add_accessory(self, accessory):
        if not hasattr(self, 'accessories'):
            self.accessories = []
        self.accessories.append(accessory)
        self.total_cost = self.calculate_total()
        logger.debug(
            f"Добавлен аксессуар {accessory.name} к аренде {self.rental_id}. "
            f"Новая стоимость: {self.total_cost}")

    def remove_accessory(self, accessory):
        self.accessories.remove(accessory)
        self.total_cost = self.calculate_total()
        logger.debug(
            f"Удален аксессуар {accessory.name} из аренды {self.rental_id}. "
            f"Новая стоимость: {self.total_cost}"
        )

    def calculate_total(self):
        if not hasattr(self, 'accessories'):
            self.accessories = []
        days = (self.end_date - self.start_date).days
        summa = self.equipment.calculate_rental_cost(days)
        for el in self.accessories:
            summa += el.calculate_rental_cost(days)
        logger.debug(f"Расчет стоимости аренды {self.rental_id}: {summa} за {days} дней")
        return summa

    def to_dict(self):
        logger.debug(f"Преобразование аренды {self.rental_id} в словарь")
        return {
            'rental_id': self.rental_id,
            'customer': self.customer_info.to_dict(),
            'equipment': self.equipment.to_dict(),
            'start_date': self.start_date.isoformat(),
            'end_date': self.end_date.isoformat(),
            'total_cost': self.total_cost
        }

    @classmethod
    def from_dict(cls, data):
        logger.debug(f"Создание аренды из словаря: {data}")
        customer = Customer.from_dict(data['customer'])
        equipment = Equipment.from_dict(data['equipment'])
        start_date = datetime.fromisoformat(data['start_date'])
        end_date = datetime.fromisoformat(data['end_date'])
        return cls(data['rental_id'], customer, equipment, start_date, end_date)

    def __str__(self):
        ac = [a.name for a in self.accessories]
        return (f"Клиент: {self.customer_info.name}\n"
                f"Оборудование: {self.equipment.name}\n"
                f"Аксессуары: {ac}\n"
                f"Период: {self.start_date} - {self.end_date}\n"
                f"Стоимость: {self.total_cost}")
```

```python
import abc
from order import Rental
from chain import Operator, Manager, Admin, check_permissions
from error import InvalidEquipmentError, RentalNotFoundError, PermissionDeniedError
from equipment import EquipmentFactory, logger
from datetime import datetime

class LogginMixin:
    def log_action(self, s):
        logger.info(f"[{datetime.now()}] {s}")

class NotificationMixin:
    def send_notification(self, client, mes):
        logger.info(f"[{datetime.now()}] Уведомление для {client}: {mes}")

class Rentabel(abc.ABC):
    @abc.abstractmethod
    def rent_equipment(self, customer, equipment, start_date, end_date):
        pass

class Reportable(abc.ABC):
    @abc.abstractmethod
    def generate_report(self):
        pass

class RentalService(Rentabel, Reportable, LogginMixin, NotificationMixin):
    rentals = []
    def __init__(self):
        self.handler = Operator(Manager(Admin()))
        self.online = OnlineRentalProcess()
        self.offline = OfflineRentalProcess()
        logger.info("Сервис пренды создан")

    @check_permissions("user")
    def rent_equipment(self, customer, data, start_date, end_date, is_online = True, **kwargs):
        try:
            logger.debug(f"Создание оборудования из данных: {data}")
            equipment = EquipmentFactory.create_equipment(data)
        except InvalidEquipmentError as e:
            logger.error(f"Ошибка создания оборудования: {e.mes}")
            raise InvalidEquipmentError(data, e.mes)

        if is_online:
            logger.debug("Испоьзуется онлайн процесс аренды")
            process = self.online
        else:
            logger.debug("Используется офлайн процесс аренды")
            process = self.offline
        rental = process.rent_equipment(customer, equipment, start_date, end_date)
        self.log_action(f"Оборудование {equipment.name} арендовано клиентом {customer.name}")
        self.send_notification(customer.email, f"Ваш заказ готово к выдаче")
        return rental

    def find_rental(self, rental_id):
        for rental in self.rentals:
            if rental.rental_id == rental_id:
                logger.debug(f"Аренда {rental_id} найдена")
                return rental
        logger.error(f"Аренда с ID {rental_id} не найдена")
        raise RentalNotFoundError(f"Аренда с ID {rental_id} не найдена")

    @check_permissions("manager")
    def change_request(self, request):
        try:
            logger.debug(f"Обработка запроса на изменение аренды {request.rental.rental_id}")
            return self.handler.handle(request)
        except PermissionDeniedError as e:
            logger.error(f"Ошибка прав доступа при изменении аренды: {e.mes}")
            raise PermissionDeniedError(e.mes)

    def generate_report(self):
        logger.info("Генерация отчета по аренде")
        report = "Отчет по аренде:\n"
        for rental in self.rentals:
            report += str(rental) + "\n"
        return report

class RentalProcess(abc.ABC):
    def rent_equipment(self, customer, equipment, start_date, end_date):
        try:
            logger.debug(f"Проверка прав клиента {customer.name} на аренду")
            customer.can_rent()
        except PermissionDeniedError as e:
            logger.error(f"Ошибка прав доступа: {e.mes}")
            raise PermissionDeniedError(e.mes)

        logger.debug(f"Проверка доступности оборудования {equipment._name}")
        self.check_availability(equipment)
        logger.debug(f"Создание аренды для оборудования {equipment._name}")
        rental = self.create_rental(customer, equipment, start_date, end_date)
        logger.debug(f"Подтверждение аренды {rental.rental_id}")
        self.confirm_rental(rental)
        logger.info(f"Успешно создана аренда {rental.rental_id}")
        return rental

    @abc.abstractmethod
    def check_availability(self, equipment):
        pass

    @abc.abstractmethod
    def create_rental(self, customer, equipment, start_date, end_date):
        pass

    @abc.abstractmethod
    def confirm_rental(self, rental):
        pass

class OnlineRentalProcess(RentalProcess):
    def check_availability(self, equipment):
        if not equipment.is_available:
            logger.error(f"Оборудование {equipment._name} недоступно для онлайн аренды")
            raise ValueError(f"Оборудование {equipment._name} недоступно для аренды онлайн")
        logger.debug(f"Оборудование {equipment._name} доступно для онлайн аренды")

    def create_rental(self, customer, equipment, start_date, end_date):
        equipment.is_available = False
        logger.debug(f"Создана онлайн аренда: ID {len(RentalService.rentals) + 1}")
        return Rental(len(RentalService.rentals) + 1, customer, equipment, start_date, end_date)

    def confirm_rental(self, rental):
        RentalService.rentals.append(rental)
        logger.info(f"Онлайн аренда {rental.rental_id} подтверждена")


class OfflineRentalProcess(RentalProcess):
    def check_availability(self, equipment):
        if not equipment.is_available:
            logger.error(f"Оборудование {equipment._name} недоступно для офлайн аренды")
            raise ValueError(f"Оборудование {equipment._name} недоступно для аренды офлайн")
        logger.debug(f"Оборудование {equipment._name} доступно для офлайн аренды")

    def create_rental(self, customer, equipment, start_date, end_date):
        equipment.is_available = False
        logger.debug(f"Создана офлайн аренда: ID {len(RentalService.rentals) + 1}")
        return Rental(len(RentalService.rentals) + 1, customer, equipment, start_date, end_date)

    def confirm_rental(self, rental):
        RentalService.rentals.append(rental)
        logger.info(f"Офлайн аренда {rental.rental_id} подтверждена")
```

```python
import abc
from error import PermissionDeniedError
from equipment import logger

def check_permissions(*roles):
    def decorator(func):
        def wrapper(*args, **kwargs):
            user = kwargs.get("user_role")
            if user not in roles:
                logger.warning(f"Попытка доступа без прав: пользователь {user}")
                raise PermissionDeniedError("Нет прав доступа")
            logger.debug(f"Доступ разрешен для пользователя {user}")
            return func(*args, **kwargs)
        return wrapper
    return decorator


ChangeRequest = ('ChangeRequest', ['rental', 'new_dates', 'new_equipment', 'new_cost'])

class Handler(abc.ABC):
    def __inti__(self, successor=None):
        self._successor = successor
        logger.debug(f"Создан обработчик {self.__class__.__name__}")

    @abc.abstractmethod
    def handle(self, request):
        pass


class Operator(Handler):
    def __init__(self, successor=None):
        self._successor = successor
    def handle(self, request):
        if request.new_dates:
            logger.info(f"Оператор изменяет даты аренды {request.rental.rental_id}")
            request.rental.start_date = request.new_dates[0]
            request.rental.end_date = request.new_dates[1]
            request.rental.total_cost = request.rental.calculate_total()
            return True
        if self._successor:
            logger.debug("Оператор передает запрос админу")
            return self._successor.handle(request)
        return False

class Manager(Handler):
    def __init__(self, successor=None):
        self._successor = successor
    def handle(self, request):
        if request.new_cost:
            logger.info(f"Менеджер изменят стоимость аренды {request.rental.rental_id}")
            request.rental.total_cost = request.new_cost
            return True
        if self._successor:
            logger.debug("Менеджер передает запрос админу")
            return self._successor.handle(request)
        return False

class Admin(Handler):
    def __init__(self, successor=None):
        self._successor = successor
    def handle(self, request):
        if request.new_equipment:
            logger.info(f"Админ изменяет оборудование в аренде {request.rental.rental_id}")
            request.rental.equipment = request.new_equipment
            request.rental.total_cost = request.rental.calculate_total()
            return True
        return False
```

файл make_json.py

```python
from equipment import Equipment, logger
from order import Rental
import json
import logging
from logging.handlers import RotatingFileHandler

class MakeJson:
    @staticmethod
    def save_data(filename, equipment, rentals):
        data = {
            'equipment': [eq.to_dict() for eq in equipment],
            'rentals': [rental.to_dict() for rental in rentals]
        }
        with open(filename, 'w') as f:
            json.dump(data, f, ident = 4, default = str)
        logger.info(f"Данные сохранены в файл: {filename}")

    @staticmethod
    def load_data(filename):
        with open(filename, 'r') as f:
            data = json.load(f)
        equipment = [Equipment.from_dict(eq) for eq in data['equipment']]
        rentals = [Rental.from_dict(rental) for rental in data['rentals']]
        logger.info(f"Данные загружены из файла: {filename}")
        return {'equipment': equipment, 'rentals': rentals}

def setup_logging():
    logger = logging.getLogger()
    logger.setLevel(logging.DEBUG)

    formatter = logging.Formatter('%(actime)s - %(name)s - %(levelname)s - %(message)s')
    file_handler = RotatingFileHandler('app.log', maxBytes=1024*1024, backupCount=5)
    file_handler.setFormatter(formatter)
    logger.addHandler(file_handler)

    console_handler = logging.StreamHandler()
    console_handler.setFormatter(formatter)
    logger.addHandler(console_handler)

setup_logging()
```

```python
class InvalidEquipmentError(Exception):
    def __init__(self, data, mes = "Некорректные данные оборудования"):
        self.data = data
        self.mes = mes
        super().__init__(self.mes)

    @classmethod
    def validate(cls, data):
        mas = ['equipment_type', 'equipment_id', 'name', 'condition', 'daily_rate']
        if not all(field in data for field in mas):
            raise cls(data, "Отсутствуют обязательные поля")

        if data['equipment_type'] not in ['Camera', 'Lens', 'Accessory']:
            raise cls(data, "Некорректный тип оборудования")

        if not isinstance(data['equipment_id'], int) or data['equipment_id'] <= 0:
            raise cls(data, "Некорректный идентификатор оборудования")

        if not isinstance(data['name'], str):
            raise cls(data, "Некорректное название оборудования")

        if not isinstance(data['condition'], str):
            raise cls(data, "Некорректное состояние оборудования")

        if not isinstance(data['daily_rate'], (int, float)) or data['daily_rate'] <= 0:
            raise cls(data, "Некорректная стоимость аренды")

        if data['equipment_type'] == 'Camera' and 'sensor_type' not in data:
            raise cls(data, "Отсутствует тип сенсора для камеры")

        if data['equipment_type'] == 'Lens' and 'focal_length' not in data:
            raise cls(data, "Отсутствует фокусное расстояние для объектива")

        if data['equipment_type'] == 'Accessory' and 'compatibility' not in data:
            raise cls(data, "Отсутсвует совместимость для аксессуара")

class PermissionDeniedError(Exception):
    def __init__(self, mes = "У пользователя нет прав доступа"):
        self.mes = mes
        super().__init__(self.mes)

class RentalNotFoundError(Exception):
    def __init__(self, mes = 'Аренда не найлена'):
        self.mes = mes
        super().__init__(self.mes)
```
